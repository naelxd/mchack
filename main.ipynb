```python
registrations = new_data['create_date'].value_counts().sort_index(ascending=False).to_dict()
```

```python
registrations
```

```python
research = {}
for key, value in registrations.items():
    year, *_ = key.split('-')
    if year not in research.keys():
        research[year] = 0
    research[year] += int(value)
```
